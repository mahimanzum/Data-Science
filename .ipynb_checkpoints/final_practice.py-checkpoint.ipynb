```python
import pandas
import cv2
import numpy as np
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
labels ='C:/Users/HP/Desktop/Data Science/data/shuffled_labels.csv'
df = pandas.read_csv(labels)
rows = df.iterrows()
x_temp = []
y_temp = []
for row in rows:
    path = row[1][0]
    img_cls = row[1][1]
    img = cv2.imread(path)
    #print(img.shape)
    x_temp.append(img)
    y_temp.append(img_cls)
    #print(row.shape)
encoder = LabelEncoder()
encoder.fit(y_temp)
encoded_Y = encoder.transform(y_temp)
Y = np_utils.to_categorical(y_temp) 
np.savez('C:/Users/HP/Desktop/Data Science/data/data.npz',X_train = x_temp, Y_train = Y)
```
